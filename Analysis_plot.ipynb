```python
import matplotlib.pyplot as plt
import numpy as np

# Data extracted from the image
bins = [(-0.4, 7.6), (7.6, 15.6), (15.6, 23.6), (23.6, 31.6), (31.6, 39.6), (39.6, 47.6), (47.6, 55.6)]
pdf_hits = [3046, 357, 162, 65, 11, 6, 9]
cdf_hits = [3046, 3403, 3565, 3630, 3641, 3647, 3656]

# Summary statistics
count = 3656
mean = 5.154
std_dev = 6.208
min_val = 0.039
max_val = 55.273
total_sum = 18841.44

# Calculate bin centers for plotting
bin_centers = [(start + end) / 2 for start, end in bins]

# Create figure and axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot PDF (hits) as bar chart
ax1.bar(bin_centers, pdf_hits, width=5, color='skyblue', label='PDF (Hits)', alpha=0.7, hatch='/')
ax1.set_xlabel('Response Time Intervals')
ax1.set_ylabel('PDF (Hits)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Enable grid for the PDF (hits) axis
ax1.grid(True, linestyle='--', color='gray', alpha=0.5)

# Plot CDF as line chart on the same axis
ax2 = ax1.twinx()
ax2.plot(bin_centers, cdf_hits, color='orange', marker='o', linestyle='-', label='CDF (Cumulative Hits)')
ax2.set_ylabel('CDF (Cumulative Hits)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Enable grid for the CDF axis as well
ax2.grid(True, linestyle='--', color='gray', alpha=0.3)

# Add summary statistics text box
stats_text = (
    f"Count: {count}\nMean: {mean:.3f}\nMin: {min_val}\nMax: {max_val}\n"
    f"Deviation: {std_dev:.3f}\nSum: {total_sum:.2f}"
)
plt.text(0.7, 0.7, stats_text, transform=ax1.transAxes, fontsize=10,
         bbox=dict(facecolor='lightyellow', edgecolor='black', boxstyle='round,pad=0.5'))

# Add legends
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Show plot
plt.title("Response Times Distribution")
plt.tight_layout()
plt.savefig("Response time from orignal data set.pdf", format="pdf", dpi=300)
plt.show()
```
